```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import interact, interact_manual 
```

```python
def draw(a,b):
    
    p = 5 + 4*1j
    z = 2*a + 3*b*1j
    
    alist = [p]
    for i in range(100):
        p = p * z
        alist.append(p)
    
    ax = plt.gca()
    ax.set_aspect("equal")
    
    x = np.real(alist)
    y = np.imag(alist)
    plt.plot(x,y)
```

```python
draw(0.1, 0.2)
```

```python
interact(draw, a =(0.001, 10), b = (0.002, 20))
```

```python
interact_manual(draw, a =(0.001, 10), b = (0.002, 20))
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from ipywidgets import interact, interact_manual
```

```python
sp.init_printing()
```

```python
I = sp.I
```

```python
x, y = sp.symbols('x,y', real = True)
```

```python
z = x + I * y
```

```python
x = np.linspace(-50, 50, 1000)
y = np.zeros(1000)+1    


plt.subplot(1,2,1)
plt.scatter(x, y)
plt.title("Z-plane")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.grid(linestyle='--')


Z = x + I * y
Z1 = 1/Z * (np.cos(x))
A1 = []
A2 = []
for i in Z1:
    A1.append(sp.re(i))
    A2.append(sp.im(i))
plt.subplot(1,2,2)
plt.scatter(A1, A2)
plt.title("W-plane")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.grid(linestyle='--')



plt.tight_layout()
```

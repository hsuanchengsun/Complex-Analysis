```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from ipywidgets import interact, interact_manual
```

```python
sp.init_printing()
```

```python
I = sp.I
```

```python
x, y = sp.symbols('x,y', real = True)
```

```python
z = x + I * y
```

```python
def equ1(a, b):
    f = sp.Eq(abs(z)**2 - 2 * a**2 * abs(z)**2, b**3 - a**4  )
    sp.plot_implicit(f)
```

```python
equ1(10, 10)
```

```python
interact(equ1, a = (0, 10), b = (0, 10))
```

```python
def equ2(a = 10, b = 10):
    g = sp.Eq(sp.im(z)**4 - sp.re(z)**4 + a * sp.im(z)**2 + b * sp.re(z)**2)
    
    sp.plot_implicit(g)
```

```python
sp.plot_implicit(g)
```

```python
interact(equ2, a = (-5, 5), b = (-5, 5))
```

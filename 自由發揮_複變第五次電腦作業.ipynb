### --------以下進入正題--------

#### 上課例題： $ S=\{  x+iy ~|~ -\frac{\pi}{2} \le x \le \frac{\pi}{2} , 0 \le y   \} ~{\rm Find} ~f(S) ~\rm{where} ~f(z)=sin(z)$ 

```python
x = np.linspace(-pi/2,pi/2,20) #設定x, y範圍
y = np.linspace(0,1,20)
X, Y = np.meshgrid(x,y) #變成下面那樣，要注意它會是x*y個點，所以前面不要給太多點

plt.scatter(X, Y)
plt.title("Z-plane")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.grid(linestyle='--')
```

```python
#用了meshgrid之後我發現對複數會不太友善，要做一點轉換
z = []
for i in range(20):
    for j in range(20):
        z.append(X[i][j] + I*Y[i][j])

#收集經過轉換之後每一點的polar coord        
theta = []
r = []
for i in z:
    a = cmath.polar(sp.sin(i))[1]
    b = cmath.polar(sp.sin(i))[0]
    theta.append(a)
    r.append(b)

    
#收集x, y座標    
s = []
t = []
for i in range(len(z)):    
    s.append(r[i]*np.cos(theta[i]))
    t.append(r[i]*np.sin(theta[i]))

plt.scatter(s, t)
plt.title("W-plane")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.grid(linestyle='--')
```

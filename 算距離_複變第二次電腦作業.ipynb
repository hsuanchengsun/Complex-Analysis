```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interact_manual 
```

```python
def equ(a, b, r):
    
    theta = np.linspace(0, 2*np.pi, 100)

    x = r*np.cos(theta) + a
    y = r*np.sin(theta) + b

    fig, ax = plt.subplots(1)

    ax.plot(x, y)
    ax.set_aspect("equal")

    plt.xlim(a-r-10, a+r+10)
    plt.ylim(b-r-10, b+r+10)
    
    plt.scatter(a, b, c = "r")

    plt.grid(linestyle='--')
```

```python
equ(1, 2, 10)
```

```python
def inequ(a, b, r):
    
    theta = np.linspace(0, 2*np.pi, 100)
    
    x = r*np.cos(theta) + a
    y = r*np.sin(theta) + b

    fig, ax = plt.subplots(1)

    ax.plot(x, y)

    plt.xlim(a-r-10, a+r+10)
    plt.ylim(b-r-10, b+r+10)
    
    plt.fill_between(x, y)
    plt.scatter(a, b, c = "r")

    plt.grid(linestyle='--')
```

```python
inequ(1, 2, 10)
```

```python
interact(equ, a = (-10, 10), b = (-10, 10), r = (0, 10))
```

```python
interact(inequ, a = (-10, 10), b = (-10, 10), r = (0, 10))
```
